# LINEAR AND NON-LINEAR LINES

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,3,100)
y = -1*(x**2) + x*3 - 1
plt.plot(x,y)
```

### Slope of a linear function

```python
def cal_slope(x1,x2,y1,y2):
    m = (y2-y1)/(x2-x1)
    print("Slope is",m)
    
case_one = cal_slope(1,2,5,13)
case_two = cal_slope(0,-1,13,-2)
```

### Non-linear functions, Secant Lines

```python
import seaborn
seaborn.set(style="darkgrid")

def secant_line(x_values):
    x = np.linspace(-20,30,100)
    y = -1*(x**2) + x*3 - 1
    plt.plot(x,y)
    
    x_0 = x_values[0]
    x_1 = x_values[1]
    y_0 = -1*(x_0**2) + x_0*3 - 1
    y_1 = -1*(x_1**2) + x_1*3 - 1
    m = (y_1 - y_0) / (x_1 - x_0)
    b = y_1 - m*x_1
    
    y_secant = x*m + b
    plt.plot(x, y_secant, c='green')
    plt.show()
    
secant_line([-10,6])
secant_line([2,10])
secant_line([-5,6])
```
