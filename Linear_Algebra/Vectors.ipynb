# Vectors

```python
import numpy as np

vector1 = np.asarray([1,2,3,4])
vector2 = np.asarray([1,1,1,1])
### adding 2 vectors
vector_1_2 = vector1 + vector2
vector_1_2
```

```python
### multiplying and dividing vectors by scalers

vector3 = vector1//2
print(vector3)

vector4 = vector2*2
print(vector4)
```

### plotting vectors

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

X = [0,1] ## x coordinate origin
Y = [0,2] ## y coordinate origin
U = [1,3] ## distance traveled in x direction
V = [2,2] ## distance travelled in y direction

plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)

plt.xlim([0,6])
plt.ylim([0,6])
plt.show()



```

```python
plt.quiver([0,1,0], [0,2,0], [1,3,4], [2,2,4], angles='xy', scale_units='xy', scale=1)
plt.xlim([0,6])
plt.ylim([0,6])
plt.show()
```

```python
### Vector Length
```

```python
X = [0,2,0]
Y = [0,0,0]
U = [2,0,2]
V = [0,3,3]

plt.quiver(X,Y,U,V, angles='xy', scale_units='xy', scale=1)
plt.xlim([0,6])
plt.ylim([0,6])
plt.show()

vector_length = (2**2 + 3**2)**0.5
vector_length
```

### Dot Product

```python
### tells us how much of one vector is pointing towards the other vector(direction)

###orthogonal vectors
X = [0,0]
Y = [0,0]
U = [1,-1]
V = [1,1]

plt.quiver(X,Y,U,V, angles='xy', scale_units='xy', scale=1)
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.show()

dot = 2*5 + 3*7 + 5*8 + 6*9  ## Dot Product for vectors [2,3,5,6] and [5,7,8,9]
print(dot)
```

### Matrix Multiplication

```python
matrix1 = np.asarray([[1,2],[3,4],[5,6]])
matrix2 = np.asarray(([1,2],[2,3]))

matrix3 = np.dot(matrix1,matrix2)
print(matrix3)
```
